```python
# Filter for BJP candidates
bjp_candidates = df[df['party'] == 'BJP']

# Sort by criminal cases and get the top 10
top_10_bjp_criminal = bjp_candidates.sort_values('criminal_cases', ascending=False).head(10)

# Display the resulting DataFrame
display(top_10_bjp_criminal)
```

```python
plt.figure(figsize=(10, 6))
plt.stem(top_10_bjp_criminal['candidate'], top_10_bjp_criminal['criminal_cases'])
plt.xticks(rotation=90)
plt.xlabel('Candidate')
plt.ylabel('Number of Criminal Cases')
plt.title('Top BJP Candidates by Criminal Cases')
plt.tight_layout()
plt.show()
```

```python
average_assets_by_education = df.groupby('education')['actual_assets'].mean().sort_values(ascending=False)

fig = px.bar(average_assets_by_education,
             x=average_assets_by_education.index,
             y=average_assets_by_education.values,
             labels={'x': 'Education Level', 'y': 'Average Actual Assets'},
             title='Average Actual Assets by Education Level')

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

criminal_cases_by_party = df.groupby('party')['criminal_cases'].sum().sort_values(ascending=False)

# Reshape the data for the heatmap (single row)
heatmap_data = criminal_cases_by_party.values.reshape(1, -1)

plt.figure(figsize=(15, 2)) # Adjust figure size for a single row
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu', xticklabels=criminal_cases_by_party.index, yticklabels=False)
plt.title("Total Criminal Cases by Political Party (Heatmap)")
plt.xlabel("Political Party")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['total_assets'], bins=30, edgecolor='black')
plt.xlabel("Total Assets")
plt.ylabel("Frequency")
plt.title("Distribution of Total Assets")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import matplotlib.pyplot as plt

average_liabilities_by_party = df.groupby('party')['liabilities'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(average_liabilities_by_party.index, average_liabilities_by_party.values)
plt.title("Average Liabilities by Political Party")
plt.xlabel("Political Party")
plt.ylabel("Average Liabilities")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
average_criminal_cases_by_education = df.groupby('education')['criminal_cases'].mean().reset_index()

fig = px.treemap(average_criminal_cases_by_education,
                 path=['education'],
                 values='criminal_cases',
                 title='Proportion of Criminal Cases by Education Level')
fig.show()
```
